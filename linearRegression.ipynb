#LINEAR REGRESSION MODEL TO PREDICT HOUSE PRICES BASED ON AREA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('Book1.csv')
```

```python
%matplotlib inline
plt.xlabel('Price')
plt.ylabel('Area')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
test = pd.read_csv('testData1.csv')
test
```

```python
predictions = reg.predict(test)
```

```python
test['price'] = predictions
test.head()
```

THE FITTED LINE USING LINEAR REGRESSION IS SHOWN BELOW

```python
%matplotlib inline
plt.xlabel('Price')
plt.ylabel('Area')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

#CANADA PER INCOME CAPITA PREDICTION

```python
data = pd.read_csv('data_1.csv')
data.head()
```

```python
model = linear_model.LinearRegression()
model.fit(data[['year']],data.income)
```

```python
%matplotlib inline
plt.ylabel('per capita income (US$)')
plt.xlabel('year')
plt.scatter(data.year,data.income,color='blue',marker='+')
plt.plot(data.year,model.predict(data[['year']]),color='red')
```

PREDICTING PER CAPITA INCOME OF THE YEAR 2020 USING OUR LINEAR REGRESSION MODEL
